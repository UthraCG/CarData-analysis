```python
#Splitting the data into Train and Test
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,
                                                y,
                                                test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

x, y = make_regression(n_features=8, n_informative=2,random_state=0, shuffle=False)
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x, y)
# RandomForestRegressor(...)

# [-8.32987858]
```

```python
print(regr.predict([[0.711340,0.866667,0.269341,0.067043,0.5,0.0,1.0,0.0]]))
```

```python
# Make predictions
y_predicted = regr.predict(xtest)
y_predicted_train = regr.predict(xtrain)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Calculate metrics
mae = mean_absolute_error(ytest, y_predicted)
mse = mean_squared_error(ytest, y_predicted)
rmse = np.sqrt(mse)
r2 = r2_score(ytest, y_predicted)
print("REPORT FOR CAR DATA SET")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R-squared: {r2}")
```
